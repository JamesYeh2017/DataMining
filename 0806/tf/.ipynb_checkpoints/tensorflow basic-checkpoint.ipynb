1. Building the computational graph.
    其實這個部分就是所謂的建立模型(Build Model)，決定整個運算流程，這也是tensorflow和其他框架最大的不同，我們必須事先定義好Graph的長相，舉下圖為例，如果想用tensorflow算 3+4=7


```python
Image(filename='./image/tf_2.png', width=400)
```

```python
node1 = tf.constant(3) # 常數3的tensor
node2 = tf.constant(4) # 常數4的tensor

node3 = tf.add(node1, node2) # 把前面創建的const tensor node1、node2相加
```


2 Running the computational graph.

```python
sess = tf.Session()
print("sess.run(node3): ",sess.run(node3)) 
```

Multiplies matrix a by matrix b, producing a * b.

這是個比較好的做法

# Placeholder 傳入值

```python
# 佔位符 , 暫時儲存變量 
#在還沒有資料的時候先佔個位子
a = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)
adder_node = a + b


# Tensorflow如果想要從外部傳入data,
# 那就需要用到tf.placeholder(),
# 然後以這種形式傳輸數據sess.run(***, feed_dict={input: **}).
```

```python
Image(filename='./image/tf_3.png', width=400)
```

```python
with tf.Session() as sess:

    print(sess.run(adder_node, {a: 3, b:4.5}))
    
    print(sess.run(adder_node, {a: [1,3], b: [2, 4]}))
  
```

# Variable 變數 (變量)

Variable在Tensorflow中的功用是甚麼呢?其實所有除了Placeholder跟Constant以外Tensor都屬於Variable，也就是剛剛做加法運算的Graph中的adder_node也屬於一個Variable，Variable就是會因為Placeholder所Input的Data不同會隨之改變，這也是在整個Model中最重要的部分。

```python
import tensorflow as tf

var = tf.Variable(1 ,name = 'counter')   
# 變量需要宣告



```

```python
print(var)
print(var.name)
#print(var) 不起作用！！
var.name
```

```python
add_operation = tf.add(var, 1) # 加法  此步没有直接计算

update_operation = tf.assign(var, add_operation)#  State 更新成 new_value
```

```python
with tf.Session() as sess:
    # once define variables, you have to initialize them by doing this
    sess.run(tf.global_variables_initializer()) # 如果定義 Variable, 就一定要 initialize

    for _ in range(3):
        sess.run(update_operation)
        print(sess.run(var))
```
